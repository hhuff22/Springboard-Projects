```python
import pandas as pd
import numpy as np
```

```python
health2018 = pd.read_csv("500_Cities__Local_Data_for_Better_Health__2018_release.csv")
dfh18 = pd.DataFrame(health2018)
```

```python
#Display dataframe
pd.options.display.float_format = '{:20}'.format
dfh18.head()
```

```python
dfh18.info()
```

```python
#Display columns
dfh18.columns
```

```python
#Filter for desired columns
dfh18 = dfh18[['Year', 'GeographicLevel', 'StateDesc', 'CityName','Category','Short_Question_Text','Measure','Data_Value_Type', 'Data_Value',
       'Low_Confidence_Limit', 'High_Confidence_Limit', 'Data_Value_Footnote', 'PopulationCount','GeoLocation','CityFIPS','TractFIPS']]

# Filter out rows at the US or City level
dfh18 = dfh18.loc[dfh18["GeographicLevel"] != "US"]
dfh18 = dfh18.loc[dfh18["GeographicLevel"] != "City"]

#Convert Tract column to integer, replace negative numbers, and display dataframe
dfh18['TractFIPS'] = dfh18['TractFIPS'].astype("int")
dfh18['TractFIPS'] = dfh18['TractFIPS'].apply(abs)
dfh18.head()
```

```python
#View unique values in the 'Measure' column
dfh18["Measure"].unique()
```

```python
#Filter values related to heart disease
dfh18 = dfh18.loc[dfh18["Measure"].isin(['High blood pressure among adults aged >=18 Years','Taking medicine for high blood pressure control among adults aged >=18 Years with high blood pressure','Coronary heart disease among adults aged >=18 Years','Cholesterol screening among adults aged >=18 Years','High cholesterol among adults aged >=18 Years who have been screened in the past 5 Years','Current smoking among adults aged >=18 Years','Diagnosed diabetes among adults aged >=18 Years','No leisure-time physical activity among adults aged >=18 Years','Obesity among adults aged >=18 Years'])]
dfh18.head()
```

```python
#Open and import the fast food csv into a dataframe
fastfood_file = pd.read_csv("fastfoodmaps_locations_2007.csv", header=None)
dfff = pd.DataFrame(fastfood_file)
dfff.head()
```

```python
#Select and rename useful columns
dfff = dfff[[1,3,4,5,7,8]]
dfff.columns = ["Name","City","State","Zip Code","Latitude","Longitude"]

# Map restaurant names to the Name column
fastfood_dict = {'m':"McDonald's",'b':"Burger King",'p':"Pizza Hut",'t':"Taco Bell",'w':"Wendy's",'k':"KFC",'j':"Jack",'h':"Hardee's",'c':"Carl's Jr.",'i':"In-N-Out"}
dfff["Name"] = dfff["Name"].map(fastfood_dict)
dfff.head()
```

```python
dfff.info()
```

```python
dfff["Zip Code"].value_counts()
```

```python
# Remove trailing digits for zip codes
zclist = []
for zc in dfff["Zip Code"]:
    if len(zc)>5:
        zclist.append(zc[:5])
    else:
        zclist.append(zc)
dfff["Zip Code"] = zclist
dfff["Zip Code"] = dfff["Zip Code"].astype("int")
```

```python
# Read zip-to-tract conversion file
tractzip = pd.read_excel("Tract-zip.xlsx")

# Extract desired columns and display the df
tractzip = tractzip[["zip","tract"]]
tractzip.head()
```

```python
tractzip.info()
```

```python
# Merge health dataframe with conversion dataframe
dfh18m = pd.merge(dfh18,tractzip,how="inner",left_on="TractFIPS",right_on="tract")
dfh18m.head()
```

```python
# Merge health dataframe with fast food location dataframe
dfh18m = pd.merge(dfh18m,dfff,how="inner",left_on="zip",right_on="Zip Code")
dfh18m.head()
```

```python
dfh18m.info()
```

```python
# Remove redundant columns and rename columns
dfh18m = dfh18m[["Year","GeographicLevel","StateDesc","CityName","Short_Question_Text","Measure","Data_Value_Type","Data_Value","Low_Confidence_Limit","High_Confidence_Limit","PopulationCount","GeoLocation","CityFIPS","TractFIPS","Name","Zip Code","Latitude","Longitude"]]
dfh18m.columns = ["Year","GeographicLevel","State","City","Category","Measure","Data_Type","Data_Value","Low_Confidence_Limit","High_Confidence_Limit","Population","Tract GeoLocation","CityFIPS","TractFIPS","Restaurant Name","Zip Code","Restaurant Latitude","Restaurant Longitude"]

# Select heart disease rows
dfh18m = dfh18m.loc[dfh18m["Category"] == "Coronary Heart Disease"]

# Remove rows with no data value
dfh18m.Data_Value = dfh18m.Data_Value.fillna("Missing")
dfh18m = dfh18m[dfh18m.Data_Value != "Missing"]

dfh18m.head()
```

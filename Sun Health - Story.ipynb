```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
health_2018 = pd.read_csv("health_2018_merged.csv")
health_2018 = health_2018[["Year","GeographicLevel","State","City","Category","Measure","Data_Type","Data_Value","Low_Confidence_Limit","High_Confidence_Limit","Population","Tract GeoLocation","CityFIPS","TractFIPS","Restaurant Name","Zip Code","Restaurant Latitude","Restaurant Longitude"]]
health_2018.head()
```

```python
tract_restaurant_counts = health_2018.groupby("TractFIPS")["Restaurant Name"].count().sort_values(ascending=False).reset_index()
tract_restaurant_counts
```

```python
tract_disease_prev = health_2018.groupby("TractFIPS")["Data_Value"].unique().reset_index()
tract_disease_prev
```

```python
disease_density = pd.merge(tract_restaurant_counts,tract_disease_prev, on="TractFIPS")
disease_density
```
